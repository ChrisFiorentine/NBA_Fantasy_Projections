```python
df.rename({
'TD': 'PassingTD',
'TD.1': 'RushingTD',
'TD.2': 'ReceivingTD',
'TD.3': 'TotalTD',
'Yds': 'PassingYDs',
'Yds.1': 'RushingYDs',
'Yds.2': 'ReceivingYDs',
'Att': 'PassingAtt',
'Att.1': 'RushingAtt'
}, axis=1, inplace=True)
```

```python
df.drop(['Rk', '2PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'PosRank', 'OvRank', 'PPR', 'Fmb', 'GS'], axis=1, inplace=True)
```

```python
rb_df = df[df['FantPos'] == 'RB']
qb_df = df[df['FantPos'] == 'QB']
wr_df = df[df['FantPos'] == 'WR']
te_df = df[df['FantPos'] == 'TE']
```

```python
rb_df.columns
```

```python
rb_df = rb_df.drop(columns = ['Cmp', 'PassingAtt',
       'PassingYDs', 'PassingTD', 'Int'])
```

```python
rb_df['FantasyPoints'] = rb_df['RushingYDs']*0.1 + rb_df['RushingTD']*6 + rb_df['Rec']
+ rb_df['ReceivingYDs']*0.1 + rb_df ['ReceivingTD']*6 - rb_df['FL']*2
```

```python
rb_df
```

```python
rb_df.sort_values('FantasyPoints', ascending = False)
```

```python
rb_df.fillna(0, inplace = True)
```

```python
lr = LinearRegression()
```

```python
drop_cols = ['Player','Tm','FantPos','FantasyPoints']
```

```python
y = rb_df.FantasyPoints
```

```python
train_rb_df = rb_df.drop(columns = drop_cols)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
      train_rb_df, y, test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train =pd.DataFrame(data=scaler.transform(X_train), columns= train_rb_df.columns)
X_test =pd.DataFrame(data=scaler.transform(X_test), columns = train_rb_df.columns)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
```
